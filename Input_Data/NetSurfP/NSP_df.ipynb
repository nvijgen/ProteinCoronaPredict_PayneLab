Create list of unique protein IDs found in proteomic data; do for each type (i.e. Intensity, Top3, iBAQ)

```python
# Abundance folder directory
file_path = '/content/drive/MyDrive/Predicting_the_Protein_Corona_Vijgen/Input_Data/Proteomics/Abundance_Files'

# Abundance .csv file to analyze - CHANGE
data = '/Bov Swiss Intensity .csv'

# Concatenate to form complete filepath
data_filepath = file_path + data

# Create data df
data_df = pd.read_csv(data_filepath)

# Create list of protein IDs
protein_ids = data_df['Entry'].unique()

# Determine how many protein IDs are in data_df
print(len(protein_ids))
```

Create URL to access fasta sequence for each protein ID found in data

```python
# Determine how many iterations for-loop does; compare with above len(protein_ids) to see if consistent
counter = 0

for protein in protein_ids:
  print(protein)
  url = 'https://www.uniprot.org/uniprot/' + protein + '.fasta'
  print(url)
  counter += 1

print(counter)
```
